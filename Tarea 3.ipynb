#Nivel I: Exploración de Datos
1. Carga de datos y mostrar 5 filas


```python
df = spark.table("samples.bakehouse.sales_customers")
display(df.limit(5))
```

2. Conteo de registros totales

```python
df.count()
```

3. Inspección de esquema

```python
df.printSchema()
```

4. Filtro geográfico: clientes de Japan

```python
display(df.filter(df.country == "Japan"))
```

5.Selección Específica: first_name, last_name, email_address

```python
display(df.select("first_name", "last_name", "email_address"))
```

6.Búsqueda de Nulos en city

```python
df.filter(df.city.isNull()).count()
```

# Nivel II: Agregaciones y Ordenamiento (Intermedio)

7.Ranking de Países: top 5 países por cantidad de clientes

```python
display(df.groupBy("country").count().orderBy("count", ascending=False).limit(5))
```

8. Análisis de Género: conteo por gender

```python
display(df.groupBy("gender").count())
```

9.Limpieza de Texto: nombres en mayúsculas

```python
display(df.selectExpr("upper(first_name) as first_name_mayus"))
```

10.Orden Alfabético: primeros 10 clientes por last_name

```python
display(df.orderBy("last_name").limit(10))
```

11.Dominios de Correo: email_address contiene 'example'

```python
display(df.filter(df.email_address.contains("example")))
```

12.Conteo de Continentes diferentes

```python
df.select("continent").distinct().count()
```

# Nivel III: Transformaciones y Lógica de Negocio (Avanzado)

13. Columna Derivada: nombre_completo

```python
from pyspark.sql.functions import concat_ws
display(df.withColumn("nombre_completo", concat_ws(" ", df.first_name, df.last_name)))
```

14.Renombramiento: postal_zip_code -> codigo_postal

```python
display(df.withColumnRenamed("postal_zip_code", "codigo_postal"))
```

15. Clasificación por Región: es_australia

```python
from pyspark.sql.functions import when
display(df.withColumn("es_australia", when(df.country == "Australia", True).otherwise(False)))
```

16.Eliminación de Datos: sin phone_number

```python
df_sin_phone = df.drop("phone_number")
display(df_sin_phone)
```

17.Cálculo de Proporción: porcentaje de clientes en 'Asia

```python
total = df.count()
asia = df.filter(df.continent == "Asia").count()
porcentaje_asia = (asia / total) * 100
```

18.Estadísticas de ID: describe customerID

```python
display(df.select("customerID").describe())
```

# Nivel IV: Integración y Eficiencia (Reto Final)

19.Join de Tablas: sales_customers y sales_transactions por customerID

```python
df_trans = spark.table("samples.bakehouse.sales_transactions")
df_join = df.join(df_trans, on="customerID", how="inner")
display(df_join)
```

20.Persistencia: guardar clientes de 'Japan' como nueva tabla

```python
df_japan = df.filter(df.country == "Japan")
df_japan.write.format("delta").mode("overwrite").saveAsTable("clientes_japon_tu_apellido")
```
